1. Print the input text without vowels using List comprehensions

```python
str = 'hai'
vowels = 'AEIOUaeiou'

for i in vowels:
  str = str.replace(i,'')
print(str)
```

2. Print only vowels from the text using list comprehension

```python
str = 'hai'
vowels = 'AEIOUaeiou'

print(''.join([i for i in str if i in vowels]))
```

4.This is a python program to print all integers that aren't divisible bt either 2 or 3 and lies between 1 and 50

```python
def syamala(a,b):
  for i in range(a,b):
    if i % 2 != 0 and i % 3 != 0:
      print(i,end = ' ')

syamala(1,51)
```

5. The program takes in an integer and prints the smallest divisor of the integer

```python
def smallest_divisor(n):
  for i in range(2, n + 1):
    if n % i == 0:
      return i

try:
  n = 6
  print(f'The smallest divisor of {n} is {smallest_divisor(n)}')
except Exception as e:
  print(e)
```

6. Write a python program to check if it is a strong number or not

```python
import math
```

```python
n = 140
for i in range(1,n+1):
  temp = i
  sum = 0
  while (temp > 0):
    rem = temp % 10
    factorial = math.factorial(rem)
    sum = sum + factorial
    temp = temp // 10
if sum == i:
  print(f'{i} is a strong number')
else:
  print(f'{i} is not a strong number')
```
